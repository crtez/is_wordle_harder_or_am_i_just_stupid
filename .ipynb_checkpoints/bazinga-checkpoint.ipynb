```python
for word in wordle:
    url = f"https://static01.nyt.com/newsgraphics/2022/2022-01-25-wordle-solver/{word.lower()}/summary.json"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Will raise an HTTPError for bad responses (4xx and 5xx)

        if response.status_code == 200:
            json_data = response.json()
            # Save the JSON data to a file
            with open(f'summary_{word.lower()}.json', 'w') as json_file:
                json.dump(json_data, json_file)
            print("JSON data downloaded and saved successfully.")
        else:
            print(f"Request returned status code: {response.status_code}")

    except requests.exceptions.HTTPError as err:
        if 400 <= response.status_code < 500:
            print(f"Client error: {response.status_code} - {err}")
        else:
            print(f"HTTP error occurred: {err}")
    except requests.exceptions.RequestException as err:
        print(f"Error during requests to {url} : {err}")
   
```

```python
import os

# Create a dictionary from the list for easy lookup
entries_dict = {entry.split(",")[0].lower(): entry for entry in entries}

# Path to the folder containing the JSON files
folder_path = "/Users/dchou/projects/is_wordle_harder_or_am_i_just_stupid"  # Update this path

# Check if the folder path exists
if not os.path.exists(folder_path):
    print(f"Folder path does not exist: {folder_path}")
else:
    # Loop through each file in the folder
    for filename in os.listdir(folder_path):
        if filename.startswith("summary_") and filename.endswith(".json"):
            # Extract the keyword from the filename
            keyword = filename[len("summary_"):-len(".json")]

            print(f"Processing file: {filename} with keyword: {keyword}")

            # Check if the keyword exists in the dictionary
            if keyword in entries_dict:
                new_filename = f"summary_{entries_dict[keyword]}.json"
                # Construct full file paths
                old_file_path = os.path.join(folder_path, filename)
                new_file_path = os.path.join(folder_path, new_filename)

                try:
                    # Rename the file
                    os.rename(old_file_path, new_file_path)
                    print(f"Renamed {filename} to {new_filename}")
                except Exception as e:
                    print(f"Error renaming {filename}: {e}")
            else:
                print(f"Keyword {keyword} not found in entries_dict")
```
