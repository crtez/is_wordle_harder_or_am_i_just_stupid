```python
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
import statistics
```

```python
folder_path = "/Users/dchou/projects/is_wordle_harder_or_am_i_just_stupid"
d = {}
for filename in os.listdir(folder_path):
    if filename.startswith("summary_") and filename.endswith(".json"):
        with open(folder_path + "/" + filename, "r") as file:
            data = json.load(file)
            d[filename] = data['average']
```

```python
d
```

```python
def extract_number(item):
    # Split the string by commas
    parts = item.split(',')
    # Extract the number part and convert it to an integer
    return int(parts[1])
```

```python
list_of_filenames = [x for x in os.listdir(folder_path) if x.endswith(".json")]
sorted_keys = sorted(list_of_filenames, key=extract_number)
sorted_keys
```

```python
word = []
normal = []
hard = []
for key in sorted_keys:
    word.append(key.split("_")[1].split(".")[0].split(",")[0])
    normal.append(d[key]['normal'])
    hard.append(d[key]['hard'])
```

```python
fig, ax1 = plt.subplots(figsize=(200, 10))

color = 'tab:blue'
ax1.set_xlabel('Word')
ax1.set_ylabel('normal', color=color)
ax1.scatter(word, normal, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.axhline(y=statistics.median(normal), color='blue', linestyle='--', label='Y1 threshold')
plt.xticks(rotation=90)

fig.tight_layout()  # to make sure the labels do not overlap
plt.show()
```
